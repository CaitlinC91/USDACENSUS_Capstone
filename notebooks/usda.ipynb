```python
import json
import requests
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import matplotlib.dates as mdates
```

```python
with open('key.json') as fi:
    credentials = json.load(fi)

api_key = credentials['api_key']
```

```python

call = "https://quickstats.nass.usda.gov/api/api_GET/?key=" + api_key + "&source_desc=CENSUS&sector_desc=CROPS&agg_level_desc=STATE&year__GE=2017&state_alpha="

res_2017 = []
states = ['AL','AK','AZ','AR','CA','CO','CT','DE','FL','GA','HI', 'ID', 'IL', 'IN', 'IA', 'KS', 'KY', 'IA','ME', 'MD','MA', 'MI','MN', 'MS','MO', 'MT','NE','NV','NH', 'NJ','NM', 'NY','NC', 'ND','OH', 'OK','OR', 'PA','RI','SC','SD', 'TN','TX','UT','VT','VA','WA','WV','WI','WY']  # List of states to iterate over

for state in states:
    endpoint = call + state  
    result = requests.get(endpoint)  
    res_2017.append(result.json()) 
    print(result)
```

```python
res_2017[0]['data']
```

```python
#This makes a usable DF. Need to create loop to run all 50 indexes
useable = pd.json_normalize(res_2017[0]['data'])
```

```python

total_2017df = pd.DataFrame()  # Initialize an empty DataFrame to store the merged results

for x in range(50):
    df = pd.json_normalize(res_2017[x]['data'])
    total_2017df = pd.concat([total_2017df, df], ignore_index=True)
    
```
